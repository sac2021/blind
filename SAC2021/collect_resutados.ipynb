```python
from pymove.core import utils as ut
import pandas as pd
import numpy as np
```

```python
files = ut.get_filenames_subdirectories('SAC/criminal_activity/')
```

```python
evaluation = [f for f in files if 'eval' in f]
```

```python
evaluation
```

```python
dfs = []
for f in evaluation:
    try:
        df_aux = pd.read_csv(f, usecols=[0,3,4,5])
        df_aux['model'] = f
        dfs.append(df_aux)
    except Exception as e:
        print("ERRO: file {}\n{}".format(f, e))

df = pd.concat(dfs)
```

```python
df
```

```python
metrics = ['acc', 'precision_macro', 'recall_macro', 'f1-macro']
dic  = {}
for c in metrics:
    dic[c] = ['mean','std']
    
df.groupby('model').agg(dic).to_csv('SAC/criminal_results.csv')
```
